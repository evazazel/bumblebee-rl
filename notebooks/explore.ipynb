```python
from bee_env.flower_world import FlowerWorldEnv
import numpy as np

social_env = FlowerWorldEnv(variance_condition="high", cue_type="social")
nonsocial_env = FlowerWorldEnv(variance_condition="high", cue_type="non_social")

print(f"Social cue salience:     {social_env.cue_salience}")
print(f"Non-social cue salience: {nonsocial_env.cue_salience}")

# Run 100 steps and count how often each env reports a perceived cue
def count_perceived_cues(env, n=200):
    obs, _ = env.reset()
    perceived = 0
    for _ in range(n):
        action = env.action_space.sample()
        obs, _, terminated, _, _ = env.step(action)
        if obs[1] == 1.0:  # cue perceived
            perceived += 1
        if terminated:
            obs, _ = env.reset()
    return perceived / n

social_rate = count_perceived_cues(social_env)
nonsocial_rate = count_perceived_cues(nonsocial_env)

print(f"\nPerceived cue rate (social):     {social_rate:.2f}")
print(f"Perceived cue rate (non-social): {nonsocial_rate:.2f}")
print(f"\nExpected: social >> non-social")
```
